```python
# Calculation of overall no-show rate
# Yes = 1 means patient didnt show up for the appoinment
# No = 0 means patient did show up for the appointment
total_appointments = cleaned_dataset['appointment_id'].nunique()
total_no_show = cleaned_dataset['no_show_binary'].value_counts().get(1)
total_no_show_rate = (total_no_show / total_appointments) * 100
print(total_appointments)
print(total_no_show)
print(total_no_show_rate)
```

```python
# A new dataframe with count of appointments on the basis of gender
gender_counts = cleaned_dataset.groupby('gender')['appointment_id'].count().reset_index()
gender_counts.columns = ['gender', 'appointments']
gender_counts['gender_share_%'] = (
    gender_counts['appointments'] / len(cleaned_dataset) * 100
).round(2)
```

```python
gender_counts
```

```python
# A new datframe with count of gender wise no_show appointments
gender_noshows = cleaned_dataset[cleaned_dataset['no_show_binary'] == 1].groupby('gender')['appointment_id'].count().reset_index()
gender_noshows.columns = ['gender', 'no_show_count']
```

```python
gender_noshows
```

```python
# merging gender_noshows and gender_counts dataframe into a new dataframe
gender_summary = gender_counts.merge(gender_noshows, on='gender')
gender_summary['no_show_rate_%'] = (
    gender_summary['no_show_count'] / gender_summary['appointments'] * 100
).round(2)
```

```python
# Statistics for waiitng days for patients who received sms and patients who didn't
sms_waiting_days = cleaned_dataset.groupby('sms_received')['waiting_days'].agg(
    count = 'count',
    mean_waiting_days = 'mean',
    median_waiting_days = 'median',
    std_waiting_days = 'std'
).round(5).reset_index()

sms_waiting_days['sms_received'] = sms_waiting_days['sms_received'].map({
    0: 'No SMS', 
    1: 'SMS Received'
})
sms_waiting_days
```

```python
# Statistics for all the combiantions of values in sms_received and waiting_days_category
sms_waiting_days_category = cleaned_dataset.groupby(['sms_received','waiting_days_category'])['waiting_days'].agg(
    appointments_count = 'count',
    mean_waiting_days = 'mean',
    median_waiting_days = 'median',
    std_waiting_days = 'std'
).round(2).reset_index()

sms_waiting_days_category['sms_received'] = sms_waiting_days_category['sms_received'].map({
    0: 'No SMS', 
    1: 'SMS Received'
})
sms_waiting_days_category
```

```python
# A new dataframe with the no_show count and percentage on the basis of waiting_days_category and sms_received?
sms_no_show = cleaned_dataset.groupby(['waiting_days_category','sms_received']).agg(
    count = ('appointment_id', 'count'),
    no_show = ('no_show_binary', 'sum')
).reset_index()
sms_no_show['no_show_%'] = (
    sms_no_show['no_show'] / sms_no_show['count'] * 100
).round(2)

sms_no_show['sms_received'] = sms_no_show['sms_received'].map({
    0: 'No SMS', 
    1: 'SMS Received'
})
sms_no_show
```
